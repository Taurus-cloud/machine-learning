### 4.7.3 如何实现数据的离散化

```python
# 1）准备数据
data = pd.Series([165,174,160,180,159,163,192,184], index=['No1:165', 'No2:174','No3:160', 'No4:180', 'No5:159', 'No6:163', 'No7:192', 'No8:184']) 
```

```python
data
```

```python
# 2）分组
# 自动分组
sr = pd.qcut(data, 3)
```

```python
type(sr)
```

```python
sr
```

```python
sr.value_counts()
```

```python
# 3）转换成one-hot编码
pd.get_dummies(sr, prefix="height")
```

```python
# 自定义分组
bins = [150, 165, 180, 195]
sr = pd.cut(data, bins)
```

```python
sr
```

```python
sr.value_counts()
```

```python
# get_dummies
pd.get_dummies(sr, prefix="身高")
```

### pd.merge合并

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                        'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                        'key2': ['K0', 'K0', 'K0', 'K0'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, how="inner", on=["key1", "key2"])
```

```python
pd.merge(left, right, how="left", on=["key1", "key2"])
```

```python
pd.merge(left, right, how="outer", on=["key1", "key2"])
```

### 4.9.2 使用crosstab(交叉表)实现

```python
# 星期数据以及涨跌幅是好是坏数据
# pd.crosstab(星期数据列, 涨跌幅数据列)
# 准备星期数据列
stock.index
```

```python
# pandas日期类型
date = pd.to_datetime(stock.index)
```

```python
date
```

```python
stock["week"] = date.weekday
```

```python
date.weekday
```

```python
stock
```

```python
# 准备涨跌幅数据列
stock["pona"] = np.where(stock["p_change"] > 0, 1, 0)
```

```python
stock.head()
```

```python
# 交叉表
data = pd.crosstab(stock["week"], stock["pona"])
```

```python
data
```

```python
data.sum(axis=1)
```

```python
data.div(data.sum(axis=1), axis=0).plot(kind="bar", stacked=True)
```

```python
data.div(data.sum(axis=1), axis=0)
```

```python
# 透视表操作
stock.pivot_table(["pona"], index=["week"])
```

### 问题3：对于这一组电影数据，如果我们希望统计电影分类(genre)的情况，应该如何处理数据？

```python
movie
```

```python
# 先统计电影类别都有哪些
movie_genre = [i.split(",") for i in movie["Genre"]]
```

```python
movie_genre
```

```python
movie_class = np.unique([j for i in  movie_genre for j in i])
```

```python
len(movie_class)
```

```python
movie
```

```python
# 统计每个类别有几个电影
count = pd.DataFrame(np.zeros(shape=[1000, 20], dtype="int32"), columns=movie_class)
```

```python
count.head()
```

```python
# 计数填表
for i in range(1000):
    count.loc[i, movie_genre[i]] = 1
```

```python
count.head()
```

```python
count.sum(axis=0).sort_values(ascending=False).plot(kind="bar", figsize=(20, 9), fontsize=20, colormap="cool")
```
